```python
import pandas as pd
import numpy as np
import pickle

FEATURES_PATH = "../data/processed/features_scaled.npy"
SONGS_PATH = "../data/processed/songs_with_emotions.csv"
MODEL_PATH = "../data/processed/kmeans_model.pkl"

X_scaled = np.load(FEATURES_PATH)
songs = pd.read_csv(SONGS_PATH)

with open(MODEL_PATH, "rb") as f:
    kmeans = pickle.load(f)
```

```python
from sklearn.metrics.pairwise import euclidean_distances

def recommend_song(emotion, pool_size=20):
    cluster_id = songs[songs["emotion"] == emotion]["cluster"].iloc[0]

    cluster_indices = songs[songs["cluster"] == cluster_id].index
    X_cluster = X_scaled[cluster_indices]

    centroid = kmeans.cluster_centers_[cluster_id].reshape(1, -1)
    distances = euclidean_distances(X_cluster, centroid).flatten()

    top_indices = cluster_indices[np.argsort(distances)[:pool_size]]

    chosen_index = np.random.choice(top_indices)

    return songs.loc[chosen_index, ["name", "artists", "emotion"]]

```

```python
recommend_song("joy")
```
