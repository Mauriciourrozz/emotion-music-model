```python
import pandas as pd

DATASET_PATH = "../data/raw/spotify_dataset_1921_2020.csv"

df = pd.read_csv(DATASET_PATH)

df.shape
```

```python
FEATURE_COLUMNS = [
    "valence",
    "energy",
    "danceability",
    "acousticness",
    "instrumentalness",
    "liveness",
    "speechiness",
    "tempo",
    "loudness"
]
```

```python
META_COLUMNS = [
    "name",
    "artists",
    "year"
]
```

```python
X = df[FEATURE_COLUMNS]
metadata = df[META_COLUMNS]

X.shape, metadata.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled.shape
```

```python
import numpy as np

PROCESSED_FEATURES_PATH = "../data/processed/features_scaled.npy"
np.save(PROCESSED_FEATURES_PATH, X_scaled)
```

```python
PROCESSED_META_PATH = "../data/processed/metadata.csv"
metadata.to_csv(PROCESSED_META_PATH, index=False)
```
