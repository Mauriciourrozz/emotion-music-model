```python
import pandas as pd
import numpy as np

RAW_PATH = "../data/raw/spotify_dataset_1921_2020.csv"
CLUSTERED_META_PATH = "../data/processed/metadata_with_clusters.csv"

df_raw = pd.read_csv(RAW_PATH)
metadata = pd.read_csv(CLUSTERED_META_PATH)
```

```python
FEATURE_COLUMNS = [
    "valence",
    "energy",
    "danceability",
    "acousticness",
    "instrumentalness",
    "liveness",
    "speechiness",
    "tempo",
    "loudness"
]

df = df_raw[FEATURE_COLUMNS].copy()
df["cluster"] = metadata["cluster"]
```

```python
cluster_summary = df.groupby("cluster").mean()
cluster_summary
```

```python
EMOTION_MAP = {
    0: "anger",
    1: "fear",
    2: "sadness",
    3: "joy"
}

metadata["emotion"] = metadata["cluster"].map(EMOTION_MAP)
metadata.head()
```

```python
FINAL_PATH = "../data/processed/songs_with_emotions.csv"
metadata.to_csv(FINAL_PATH, index=False)
```
